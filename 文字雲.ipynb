```python
import jieba
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from collections import Counter
import ipywidgets as widgets
from IPython.display import display, clear_output

# --- 1. 準備停用詞 (Stop Words) ---
# 這些字不會出現在雲圖中
STOP_WORDS = {'的', '了', '是', '在', '有', '就', '不', '也', '和', '與', '或', '之', 'is', 'a', 'the', 'and'}

def run_wordcloud_logic(b):
    with output_area:
        clear_output()
        raw_text = text_area.value.strip()

        if not raw_text:
            print("⚠️ 錯誤：請先輸入文字再產生文字雲！")
            return

        # --- 2. 運用 Hash 技巧計數 ---
        # 這裡使用 jieba 斷詞處理中文，再存入字典 (Hash Table)
        words = jieba.cut(raw_text)
        word_counts = {} # 這就是一個 Hash Map

        for w in words:
            w = w.strip().lower()
            if len(w) > 1 and w not in STOP_WORDS: # 過濾掉單個字與無意義字
                word_counts[w] = word_counts.get(w, 0) + 1

        if not word_counts:
            print("⚠️ 過濾後沒有足夠的詞彙可以產生圖形。")
            return

        # --- 3. 繪製文字雲 ---
        wc = WordCloud(
            font_path=FONT_PATH, # 使用第一步安裝的路徑
            width=800,
            height=450,
            background_color='white',
            max_words=30, # 顯示前 n 個高頻字
            colormap='viridis' # 設定顏色風格
        ).generate_from_frequencies(word_counts)

        # 顯示圖片
        plt.figure(figsize=(12, 6))
        plt.imshow(wc, interpolation='bilinear')
        plt.axis("off")
        plt.show()

        # 存檔 (作業要求 1)
        wc.to_file("wordcloud_result.png")
        print("💾 成功！圖片已儲存為 wordcloud_result.png (點擊左側資料夾圖示可下載)")

# --- 4. 建立介面 (GUI) ---
text_area = widgets.Textarea(
    placeholder='在此貼上你想分析的文章...',
    description='輸入內容:',
    layout={'height': '200px', 'width': '95%'}
)

btn = widgets.Button(
    description='產生文字雲',
    button_style='info',
    icon='cloud'
)

output_area = widgets.Output()
btn.on_click(run_wordcloud_logic)

print("請在下方輸入文字：")
display(text_area, btn, output_area)
```
